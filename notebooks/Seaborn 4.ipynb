### Regression Plots

- regplot
- lmplot

In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.

```python
import seaborn as sns
```

```python
tips =sns.load_dataset('tips')
tips
```

```python
# axes level
# hue parameter is not available
sns.regplot(data=tips,x='total_bill',y='tip',color='Red')
```

```python
#lmplot ->only difference is you can add hue parameter
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex')
```

```python
#residual plot
#error in prediction 
sns.residplot(data=tips,x='total_bill',y='tip')
```

### Multigrid plot

#### A second way to plot facet plots -> Facetgrid


```python
# figure level -> relplot -> displot-> catplot -> lmplot
sns.catplot(data=tips,x='sex',y='total_bill',kind='violin',col='day',row='time')
```

```python
# one more way to do it is using facetgrid
# You can do easily using it
# More options
g = sns.FacetGrid(data=tips,col='day',row='time')
#it create an object which we store in g 
g.map(sns.violinplot,'sex','total_bill')
```

```python
g = sns.FacetGrid(tips, col="time", hue="sex")
g.map_dataframe(sns.scatterplot, x="total_bill", y="tip")
```

```python
g = sns.FacetGrid(tips, col="time", hue="sex")
g.map_dataframe(sns.scatterplot, x="total_bill", y="tip")
g.add_legend()
```

#### Plotting Pairwise Relationship (PairGrid vs Pairplot)

```python
iris = sns.load_dataset('iris')
iris
```

```python
# It draws the scatterplot between each possible combination of pairs
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species')
```

```python
#pair plot
g = sns.PairGrid(data=iris)
#g map
g.map(sns.scatterplot)
```

```python
# map_diag -> map_offding
g = sns.PairGrid(data=iris)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
 # you can add any type of plot
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.violinplot)   # add different types of graphplot 
g.map_offdiag(sns.scatterplot)
```

```python
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.boxplot)   
g.map_offdiag(sns.histplot)
```

```python
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.boxplot)  
g.map_offdiag(sns.kdeplot)
```

```python
#map_diag -> map_upper -> map_lower
g = sns.PairGrid(data=iris,hue='species',vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
# Here we get two many customizations 
```
