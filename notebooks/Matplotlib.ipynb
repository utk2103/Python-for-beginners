Great control with high quality output in many customizable with labels.

```python
## Plotting using matplotlib 

##plt scatter

plt.scatter(x,y,c='Blue')  #in which points are represented on graph
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('Test.jpeg') # you can save the file in whatever format you want to be
```

```python
y = x*x*x
```

```python
## plt plot
plt.plot(x,y,c='red')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d diagram')
```

```python
## plot 

plt.plot(x,y,'b*',linestyle='dashed',linewidth=4,markersize=15)
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.title('2d diagram')
```

```python
## Creating Subplots

plt.subplot(2,2,1)   ##to create a grid of subplots with 2 rows, 2 columns, and select the first subplot.
plt.plot(x,y,'r--') ## r-- represent color and dashed line 
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```

```python
# Compute the x and y coordinates for points on a sine curve
x =np.arange(0,4* np.pi, 0.1)
y =np.sin(x)
plt.title("sine wave form")
plt.plot(x,y)
plt.title('Sine graph')
```

```python
#Subplot()
# Compute the x and y coordinates for points on sine and cosine curves 
x = np.arange(0, 5 * np.pi, 0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  
   
# Set up a subplot grid that has height 2 and width 1, 
# and set the first such subplot as active. 
plt.subplot(2, 1, 1)
   
# Make the first plot 
plt.plot(x, y_sin,'r--') 
plt.title('Sine')  
   
# Set the second subplot as active, and make the second plot. 
plt.subplot(2, 1, 2) 
plt.plot(x, y_cos,'g--') 
plt.title('Cosine')  
   
# Show the figure. 
plt.show()
```

```python
## Bar plot 
x=[2,8,11]
y=[11,3,19]

x2=[3,9,11]
y2=[6,15,7]
plt.bar(x,y)
plt.bar(x2,y2,color ='g')
plt.title('Bar graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.show()
```

```python
a = np.array([22,87,5,43,56,73,55,43,54,20,52,31,27])
plt.hist(a, bins= 25)
plt.title('Histogram')
plt.show()
```

# Box plot using Matplotlib

```python
data = [np.random.normal(0, std, 100) for std in range(1,5)]

# rectangular box plot
plt.boxplot(data,vert=False,patch_artist=True);
```

```python
data = [np.random.normal(0, std, 100) for std in range(1,5)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=False);
```

# Pie chart

```python

# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```
