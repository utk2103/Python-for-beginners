## Distribution plot

• Used for univariate analysis (Analysis on only one thing)

• Used to find out the distribution

• Range of the observation

• Central Tendency

• Is the data bimodial?

• Are the outliers?

### Plots under distribution plot

• Histplot

• Kdeplot

• Rugplot


```python
# figure level-> displot
# Axis level -> Histplot -> Kdeplot -> Rugplot
```

```python
import seaborn as sns
import matplotlib.pyplot as plt     
import plotly.express as px
```

```python
tips= sns.load_dataset('tips')
tips
```

```python
# Plotting univariate histogram
sns.histplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips, x='total_bill', kind='hist')
```

```python
#thikness of bar
sns.displot(data=tips, x='total_bill', kind='hist',bins=24)
```

```python
# It's also possible to visualise the distribution of a categorical variable using the logic of histogram
# Discrete bins are automatically set for categorical variables
 
    #countplot

sns.displot(data=tips, x='day',kind='hist')
```

```python
#hue parameter
sns.displot(data=tips, x='tip',kind='hist',hue='sex')
```

```python
sns.displot(data=tips, x='tip',kind='hist',hue='sex',element='step')
```

```python
# importing the dataset
titanic =sns.load_dataset('titanic')
titanic
```

```python
sns.displot(data=titanic, x='age',kind='hist',hue='sex',element='step')
```

```python
# faceting using col and row -> not work on histplot function
sns.displot(data=tips, x='tip',kind='hist',col='sex',element='step')
```

```python
# kernel density estimation(kde plot)
# Rather than using discrete bins, a KDE plor smooths the observations with a Gaussian kernel, producing a continuous density estimates
sns.kdeplot(data=tips,x='total_bill')
```

```python
#hue
sns.kdeplot(data=tips,x='total_bill',hue='sex',fill='true')
```

```python
#to increase the size of fighure level function
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',height=10,aspect=2)
```

```python
# Rugplot

# used as a support for other graphs
# Plot marginal distribution by drawing ticks along the x and y axes.
# This function is intended to complement other plots by showing the location at individual observations in an unobtrusive way.
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

```python
# Bivariate histogram
# A bivarite histogram bins the data within rectangles that title the plot 
# and then shows the count of observations within each rectangle with the fill color
 
sns.histplot(data=tips,x='total_bill',y='tip')
```

```python
# Bivariate Kdeplot 
# a bivariate KDE plot smoothes the (x,y) observations with a 2D Gaussian
sns.kdeplot(data=tips,x='total_bill',y='tip')
```

## Matrix Plot

• Heatmap

• Clustermap


```python
# Heatmap
# Plot rectangular data as a color-encoded matrix
gap =  px.data.gapminder() 
# this is long type of data and we have to convert it into wide data so use following function 
temp_df =gap.pivot(index='country',columns='year',values='lifeExp')
```

```python
#axes level function
sns.heatmap(temp_df)
```

```python
#to change the dimension
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

```python
#annot

temp_df = gap[gap['continent']=='Europe'].pivot(index='country',columns='year',values='lifeExp')
plt.figure(figsize=(15,15))
sns.heatmap(temp_df,annot=True)
```

```python
# linewidth and cmap(color)
temp_df = gap[gap['continent']=='Europe'].pivot(index='country',columns='year',values='lifeExp')
plt.figure(figsize=(15,15))
sns.heatmap(temp_df,annot=True,linewidth=0.5,cmap='summer')
```

```python
#Clustermap

# Plot a matrix dataset as a hierarchially-clustered heatmap.

# This function requires scipy to be a available.
iris=px.data.iris()
iris
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
#dendogram ->?
# Here a ml algorithm also work known as hiercial agglomorative clustering
# here hiercial means based on degree of closeness of two data it will be plotted
```
